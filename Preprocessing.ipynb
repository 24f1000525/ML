### Handling Missing Value


SimpleImputer

```python
X=np.array([[1,2],[np.nan,3],[2,5],[3,3]])
```

```python
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(strategy='mean')
imputer.fit_transform(X)
```

```python
x_test=np.array([[1,2],[np.nan,9]])
X_test_imputed=imputer.transform(x_test)
X_test_imputed
```

KNNImputer

```python
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=2)
x_imputed=imputer.fit_transform(X)
x_imputed
```

```python
np.random.seed(42)
data=pd.DataFrame({
    'age':np.random.randint(18,60,30),
    'salary':np.random.randint(20000,100000,30),
    'experience':np.random.choice([1,2,3,4,np.nan],30),
    'education':np.random.choice(['bachelor','master','phd'],30),
    'department':np.random.choice(['hr','sales','marketing'],30),
    'city':np.random.choice(['newyork','london','paris'],30)
})
y=np.random.choice([0,1],30)
```

```python
data.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,StandardScaler
from sklearn.impute import SimpleImputer

preprocessor=ColumnTransformer(
    transformers=[
        ('imp',SimpleImputer(strategy='mean'),['experience']),
        ('ss',StandardScaler(),['age','salary']),
        ('oe',OneHotEncoder(sparse_output=False),['department','city']),
        ('ohe',OrdinalEncoder(categories=[['bachelor','master','phd']]),['education'])
    ],
    remainder='passthrough'
)
preprocessor.set_output(transform='pandas')

x_transformed=preprocessor.fit_transform(data)

x_transformed.head(10)

```
